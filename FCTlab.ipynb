```python
x = df['Current (A)']
y = df['Voltage (V)']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))   # size of figure

plt.plot(x, y)

plt.xlabel('Current (A)')
plt.ylabel('Voltage (V)')
plt.title('Current vs voltage')

plt.grid(True)

plt.show()
```

```python
df_sorted = df.sort_values(by='Current (A)')

plt.figure(figsize=(10,5))
plt.plot(df_sorted['Current (A)'], df_sorted['Voltage (V)'])
plt.xlabel('Current (A)')
plt.ylabel('Voltage (V)')
plt.title('Current vs Voltage')
plt.grid(True)
plt.show()
```
